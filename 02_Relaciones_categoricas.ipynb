# Importación de librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

# Lectura del archivo ya procesado

```python
df = pd.read_csv("data\incidentes-viales-c5-limpio.zip", sep="$", index_col=0)
```

```python
df.head()
```

# Funciones para colorear la gráfica categórica de burbujas

```python
def floatRgb(mag, cmin, cmax):
    # Normalizar a 0-1
    try: x = float(mag-cmin)/(cmax-cmin)
    except ZeroDivisionError: x = 0.5 # cmax == cmin
    azul  = min((max((4*(0.75-x), 0.)), 1.))
    rojo   = min((max((4*(x-0.25), 0.)), 1.))
    verde = min((max((4*math.fabs(x-0.5)-1., 0.)), 1.))
    return rojo, verde, azul
def rgb(mag, cmin, cmax):
    rojo, verde, azul = floatRgb(mag, cmin, cmax)
    return int(rojo*255), int(verde*255), int(azul*255)
def clamp(x): 
    return max(0, min(x, 255))
```

# Test de chi cuadrada
## Análisis de dependencia entre la variable delegación con el tipo de incidente y clasificación de alarma

### Delegación vs tipo de incidente

##### Lectura y generación de la tabla de contingencia

```python
cross_tab = df.groupby(["delegacion_inicio","incidente_c4"]).agg("count")["folio"].unstack(1).fillna(0).astype("int")
```

##### Test de chi cuadrada

```python
chi_2, p, dof, expected = chi2_contingency(cross_tab)
print('Grados de libertad = %d' % dof, '\n')

# Usando el estadístico de chi cuadrada
prob = 0.90
critical = chi2.ppf(prob, dof)
print("Usando el valor estadístico de chi cuadrada:")
print('Probabilidad = %.3f, Valor Crítico = %.3f, Chi-Cuadrada = %.3f' % (prob, critical, chi_2))
if abs(chi_2) >= critical:
    print('Ambas variables son dependientes (rechazamos la hipótesis nula)', '\n')
else:
    print('Ambas variables son independientes (no rechazamos la hipótesis nula)', '\n')
    
# Usando el p-value
alpha = 1.0 - prob
print("Usando el valor p:")
print('Significancia = %.3f, valor p = %.3f' % (alpha, p))
if p <= alpha:
    print('Ambas variables son dependientes (rechazamos la hipótesis nula)')
else:
    print('Ambas variables son independientes (no rechazamos la hipótesis nula)')
```

```python
# Valores esperados
pd.DataFrame(expected)
```

##### Visualización

```python
df_bubble_1 = cross_tab.unstack().reset_index().rename(columns={"incidente_c4":"tipo_accidente","delegacion_inicio":"lugar",0:"frecuencia"})
```

El siguiente gradiente de colores sirve para señalar la magnitud de la frecuencia de la siguiente gráfica: entre más azul claro, más baja será la frecuencia.

![title](img/gradiente_balloonplot.png)

```python
# Se crea un padding para las bolitas
df_bubble_1["padd"] = 2.5 * (df_bubble_1.frecuencia - df_bubble_1.frecuencia.min()) / (df_bubble_1.frecuencia.max() - df_bubble_1.frecuencia.min()) + 0.5
fig = plt.figure(figsize=(20,20))
# Crear el scatterplot categórico
s = plt.scatter(df_bubble_1.tipo_accidente, df_bubble_1.lugar, s = 0)
# Se grafican los datos de renglón en renglón como texto con círculos cuyos radios dependen del valor
for row in df_bubble_1.itertuples():
    col_per = rgb(row.frecuencia,cross_tab[row.tipo_accidente].min(),cross_tab[row.tipo_accidente].max())
    color = "#{0:02x}{1:02x}{2:02x}".format(clamp(col_per[0]), clamp(col_per[1]), clamp(col_per[2]))
    bbox_props = dict(boxstyle = "circle, pad = {}".format(row.padd), fc = "w", ec = color, lw = 3)
    plt.annotate(str(row.frecuencia), xy = (row.tipo_accidente, row.lugar), bbox = bbox_props, ha="center", va="center", zorder = 2, clip_on = True)
# Se grafica el grid detrás de los marcadores
plt.grid(ls = "--", zorder = 1)
# Para alinear bien las etiquetas de los ejes
fig.autofmt_xdate()
plt.xlabel("Tipo de accidente", fontsize=15, fontweight="bold")
plt.ylabel("Delegación", fontsize=15, fontweight="bold")
plt.title("Gráfica categórica de burbujas: frecuencias entre las variables de lugar y tipo de accidente", 
          fontsize=20, fontweight="bold")
plt.tight_layout()
plt.show()
fig.savefig("img/del_vs_acc.png")
```
